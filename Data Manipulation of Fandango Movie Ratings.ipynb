**Q22:** The article from 538 goes into detail about the rounding problem with the Fandango star scores.  Let's see if we can do better.  
- First create a copy of the `fandango` DataFrame.  
- Create a new column called `Fandango_Stars_Revised` that rounds the `Fandango_Ratingvalue` column to the nearest half a star.  For example, 3.2 would round to 3.0, 3.3 would round to 3.5, 3.7 would round to 3.5, 3.8 would round to 4.0, etc.  
- Select only the `FILM`, `Fandango_Ratingvalue`, and `Fandango_Stars_Revised` columns.

```python
Q22A = fandango.copy()
Q22A['Fandango_Stars_Revised'] = (Q22A['Fandango_Ratingvalue']*2).round()/2
Q22B = Q22A[['FILM', 'Fandango_Ratingvalue','Fandango_Stars_Revised']]
Q22B

#code check
# Q22B[Q22B['Fandango_Ratingvalue'] == 3.7]
Q22B[Q22B['Fandango_Ratingvalue'] == 3.8]
```

**Q23:** Calculate the sum for the `Fandango_Difference` column.

```python
Q23 = fandango['Fandango_Difference'].sum().round(2)
Q23
```

**Q24:** 
- Create a new Series called by subtracting `Fandango_Ratingvalue` from `Fandango_Stars_Revised` from the DataFrame in Q22.  
- Calculate the sum of this Series.

```python
ser = Q22A['Fandango_Stars_Revised'] - Q22A['Fandango_Ratingvalue']
Q24 = ser.sum().round(2)
Q24
```

We see that the sum for `ser` is much less than the sum of the `Fandango_Difference` column.  This makes sense as our `Fandango_Stars_Revised` column rounded to the closest half star instead of always rounding up like was done with the `Fandango_Stars` column as discussed in the 538 article.
